Для каждого алгоритма поиска подсчитать среднее (необходимо измерить 10 раз) время выполнения поиска для 10,100,1000,10000,100000 элементов. Список элементов и искомый элемент выбираются рандомно.
Алгоритмы:
- линейный поиск
- линейный поиск с break
- линейный поиск с барьером
- бинарный поиск (учитывать время вместе и без сортировкой)
- jump search (учитывать время вместе и без сортировкой)

Сформировать для всех график, где по оси х - количество элементов, а по оси у - время выполнения поиска. Графики можно построить в Excel или с помощью mathplotlib.

Для подсчета времени использовать time.perf_counter() - возвращает время с максимальной точностью. В подсчет времени не нужно включать время создание случайного массива.

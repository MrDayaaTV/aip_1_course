Для каждого алгоритма сортировки подсчитать среднее (необходимо измерить 10 раз) время выполнения сортировки для 10,100,1000,10000,100000 элементов. Список элементов выбираются рандомно.
Алгоритмы:
- пузырьковая сортировка
- сортировка выбором
- дополнительно - сортировка вставками

Сформировать для всех график, где по оси х - количество элементов, а по оси у - время выполнения сортировки. Графики можно построить в Excel или с помощью mathplotlib.

Для подсчета времени использовать time.perf_counter() - возвращает время с максимальной точностью. В подсчет времени не нужно включать время создание случайного массива.
